```python
import pandas as pd
import numpy as np
import yfinance as yf
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint, adfuller
import matplotlib.pyplot as plt
import statistics

# 1) Download daily closing prices for two assets
tickers = ['AAPL', 'MSFT']
start_date = '2018-01-01'
end_date = '2023-01-01'

df = yf.download(tickers, start=start_date, end=end_date)
df.dropna(inplace=True)  # Remove any rows with NaNs

# Separate price series for clarity
prices1 = df.Close['AAPL']
prices2 = df.Close['MSFT']
```

```python
df1 = pd.read_csv('prices_round_1_day_0.csv')
df2 = pd.read_csv('prices_round_1_day_-1.csv')
df3 = pd.read_csv('prices_round_1_day_-2.csv')
```

```python
ink_d0 = df1[df1['product'] == 'SQUID_INK']
ink_d1 = df2[df2['product'] == 'SQUID_INK']
ink_d2 = df3[df3['product'] == 'SQUID_INK']

Kelp_d0 = df1[df1['product'] == 'KELP']
kelp_d1 = df2[df2['product'] == 'KELP']
kelp_d2 = df3[df3['product'] == 'KELP']


prices1 = ink_d0.mid_price
prices1 = prices1.reset_index()
prices1 = pd.Series(prices1.mid_price)

prices2 = Kelp_d0.mid_price
prices2 = prices2.reset_index()
prices2 = pd.Series(prices2.mid_price)

spread = prices2 - prices1
```

```python
plt.figure(figsize=(12,6))
plt.plot(spread, label='Spread (prices1 - beta*prices2)')
plt.axhline(spread.mean(), color='red', linestyle='--', label='Mean')
plt.legend()
plt.title('Cointegration Spread')
plt.show()
```

```python
def linear_regression(X, y):
    X_bias = np.c_[np.ones((X.shape[0], 1)), X]
    theta = np.linalg.inv(X_bias.T.dot(X_bias)).dot(X_bias.T).dot(y)
    return(theta)
```

```python
a, b = statistics.linear_regression(spread.index,spread)
```

```python
spread_star = []
for i in range(len(spread)):
    spread_star.append(spread[i] - (a*i + b))
```

```python
corr = prices1.pct_change().corr(prices2.pct_change())
print(f"Correlation of daily returns: {corr:.2f}")
```

```python
coint_t, p_value, crit_values = coint(prices1, prices2)
print(f"Cointegration test p-value: {p_value}")
print(f"Critical values (90%, 95%, 99%): {crit_values}")
```

p-value below ~0.05 (depending on your tolerance) typically indicates you reject the null hypothesis of “no cointegration” and accept that they’re cointegrated.

The coint test also gives you the test statistic (coint_t) and critical values. If your test statistic is less than the critical value (especially at 5% or 1% level), that also indicates cointegration.

```python
X = sm.add_constant(prices2)  # e.g. try to predict prices1 from prices2
model = sm.OLS(prices1, X).fit()
beta = model.params[1]  # slope coefficient

# Spread = prices1 - beta * prices2
spread = prices1 - beta*prices2
```

```python
adf_stat, adf_pval, _, _, critical_vals, _ = adfuller(spread_star)
print(f"ADF p-value on the spread: {adf_pval}")
print(f"Critical values: {critical_vals}")
```

```python
#If the ADF test p-value is below 0.05 (or 0.01), that reinforces the spread is stationary.
```

```python
# rolling mean & std of the spread
window = 20
rolling_mean = spread.rolling(window).mean()
rolling_std = spread.rolling(window).std()

z_score = (spread - rolling_mean) / rolling_std
```

```python
z_score[z_score < -2]
```

```python
z_score[z_score > 2]
```

```python
Go long the spread (e.g. buy asset1, short asset2) if z_score < -2.
Go short the spread (short asset1, long asset2) if z_score > 2.
Close positions when the z_score crosses back toward zero.
```
